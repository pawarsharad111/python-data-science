```python
import numpy as np
import pandas as pd
```

```python
n1=np.arange(1,11)
```

```python
n1
```

```python
n2=n1**2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(n1,n2)
plt.xlabel("Numbers")
plt.ylabel("Squre")
```

```python
plt.plot(n2,n1)
plt.xlabel("Squre")
plt.ylabel("Numbers")
```

```python
plt.subplot(1,2,1)
plt.plot(n1,n2)
plt.xlabel("Number")
plt.ylabel("Square")
plt.title("Squre plot")
plt.subplot(1,2,2)
plt.xlabel("squre")
plt.ylabel("Number")
plt.plot(n2,n1)
plt.title("Squre plot")
```

```python
fig=plt.figure()

ax1=fig.add_axes([0.1,0,0.9,0.9])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(n1,n2,'g*-')
ax1.set_xlabel("number")
ax1.set_ylabel("squre")
ax1.set_title("Object Oriented plots using figure")


ax2.plot(n2,n1,'b--')
ax2.set_xlabel("Squre")
ax2.set_ylabel("number")
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,0.5,0.9])
ax2=fig.add_axes([0.6,0,0.5,0.9])
ax1.plot(n1,n2,'g*-')
ax1.set_xlabel("number")
ax1.set_ylabel("squre")
ax1.set_title("plots")

ax2.plot(n2,n1,'b--')
ax2.set_xlabel("Squre")
ax2.set_ylabel("number")
```

```python
fig, axes=plt.subplots(1,2)
axes[0].plot(n1,n2)
axes[0].set_title("squre")
axes[0].set_xlabel('Number')
axes[0].set_ylabel("squre")
axes[1].plot(n2,n1)
axes[1].set_title("Number")
```

```python
fig, axes=plt.subplots(figsize=(6,7))
axes.plot(n1,n2)
fig.savefig('abc.png')
```

```python
x=np.arange(1,11)
```

```python
fig, axes=plt.subplots()

axes.plot(x,x-1,color='blue',linewidth=2)
axes.plot(x,x,color='red',linewidth=0.12)
axes.plot(x,x+1,color='red',linewidth=0.25)
axes.plot(x,x+2,color='red',linewidth=0.50)
axes.plot(x,x+3,color='red',linewidth=0.75)
axes.plot(x,x+4,color='black',linewidth=2)

axes.plot(x,x+5,color='green',lw=3, linestyle='--')
axes.plot(x,x+6,color='green',lw=3, linestyle=':')
axes.plot(x,x+7,color='green',lw=3, ls='-.')
axes.plot(x,x-0.5,color='green',lw=3, ls='-')
axes.plot(x,x+8,color='green',lw=3, ls='-')

axes.plot(x,x+9,color='green',lw=2, ls='-.',marker='+')
axes.plot(x,x+10,color='green',lw=2, ls='--',marker='o')
axes.plot(x,x+11,color='green',lw=2, ls='-',marker='s')
axes.plot(x,x+12,color='green',lw=2, ls='--',marker='1')


```
